```python
from GmailApiPackage import authenticate, show_label_ids, get_messages, get_message_links
```

```python
service = authenticate()
print(service)
```

```python
display(show_label_ids(user_email='me', auth_var=service))
```

```python
memes_list = get_messages(user_email ='me', label_id = 'Label_2537674362841006328', auth_var = service)

print(memes_list)
```

```python
unread_list = get_messages(user_email ='me', label_id = 'UNREAD', auth_var = service)

final_list = list(set(memes_list) & set(unread_list))

print(final_list)
```

```python
extracted_urls = [get_message_links(user_email ='me', email_id = i, auth_var = service) for i in final_list]

from itertools import chain
extracted_urls = list(chain.from_iterable(extracted_urls))

display(extracted_urls)
```

```python
import webbrowser
for url in extracted_urls:
    webbrowser.open(url)
```
